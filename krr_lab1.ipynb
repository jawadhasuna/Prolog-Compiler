```python
#example fruits
from owlready2 import *

onto = get_ontology("http://example.org/fruit-ontology.owl")

with onto:

    class Fruit(Thing):
        pass

    class CitrusFruit(Fruit):
        pass

    class Berry(Fruit):
        pass

    class TropicalFruit(Fruit):
        pass

    class Color(Thing):
        pass

    class hasColor(ObjectProperty):
        domain = [Fruit]
        range = [Color]

    class hasTaste(DataProperty):
        domain = [Fruit]
        range = [str]

    Red = Color("Red")
    Yellow = Color("Yellow")
    Green = Color("Green")

    Apple = Fruit("Apple")
    Apple.hasColor = [Red]
    Apple.hasTaste = ["Sweet"]

    Lemon = CitrusFruit("Lemon")
    Lemon.hasColor = [Yellow]
    Lemon.hasTaste = ["Sour"]

    Strawberry = Berry("Strawberry")
    Strawberry.hasColor = [Red]
    Strawberry.hasTaste = ["Sweet"]

    Mango = TropicalFruit("Mango")
    Mango.hasColor = [Yellow]
    Mango.hasTaste = ["Sweet"]

print("Classes:")
for cls in onto.classes():
    print(cls)

print("\nFruits and Properties:")
for ind in onto.individuals():
    if isinstance(ind, Fruit):
        print("\nFruit:", ind.name)
        print("Color:", [c.name for c in ind.hasColor])
        print("Taste:", ind.hasTaste)

onto.save(file="fruit_ontology.owl", format="rdfxml")
```
